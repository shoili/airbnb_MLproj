```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import nltk
```

```python
data = pd.read_csv("reviews.csv")
data
```

```python
def get_sentiment(text):
#     print(type(text['comments']))
#     print(text['comments'])

    if(pd.isnull(text['comments'])):
        return 0
    return TextBlob(text['comments']).sentiment.polarity


data['sentiment'] = data.apply(lambda x: get_sentiment(x) , axis=1)
```

```python
try:
    from nltk import wordpunct_tokenize
    from nltk.corpus import stopwords
except ImportError:
    print ('[!] You need to install nltk (http://nltk.org/index.html)')



#----------------------------------------------------------------------
def _calculate_languages_ratios(text):
    """
    Calculate probability of given text to be written in several languages and
    return a dictionary that looks like {'french': 2, 'spanish': 4, 'english': 0}
    
    @param text: Text whose language want to be detected
    @type text: str
    
    @return: Dictionary with languages and unique stopwords seen in analyzed text
    @rtype: dict
    """

    languages_ratios = {}

    '''
    nltk.wordpunct_tokenize() splits all punctuations into separate tokens
    
    >>> wordpunct_tokenize("That's thirty minutes away. I'll be there in ten.")
    ['That', "'", 's', 'thirty', 'minutes', 'away', '.', 'I', "'", 'll', 'be', 'there', 'in', 'ten', '.']
    '''

    tokens = wordpunct_tokenize(text)
    words = [word.lower() for word in tokens]

    # Compute per language included in nltk number of unique stopwords appearing in analyzed text
    for language in stopwords.fileids():
        stopwords_set = set(stopwords.words(language))
        words_set = set(words)
        common_elements = words_set.intersection(stopwords_set)

        languages_ratios[language] = len(common_elements) # language "score"

    return languages_ratios


#----------------------------------------------------------------------
def detect_language(text):
    """
    Calculate probability of given text to be written in several languages and
    return the highest scored.
    
    It uses a stopwords based approach, counting how many unique stopwords
    are seen in analyzed text.
    
    @param text: Text whose language want to be detected
    @type text: str
    
    @return: Most scored language guessed
    @rtype: str
    """
    if(pd.isnull(text['comments'])):
        return "english"

    ratios = _calculate_languages_ratios(text['comments'])

    most_rated_language = max(ratios, key=ratios.get)

    return most_rated_language
```

```python
data['language'] = data.apply(lambda x: detect_language(x) , axis=1)
```

```python
def wordcount(text):
    
    if(pd.isnull(text['comments'])):
        return 0
    return len(text['comments'].split(" "))

data['wordcount'] = data.apply(lambda x:  wordcount(x)  , axis=1)

len(data[data['language']!='english'])
```

```python
def charcount(text):
    
    if(pd.isnull(text['comments'])):
        return 0
    return len(text['comments'])

data['charcount'] = data.apply(lambda x:  charcount(x) , axis=1)
```

```python
data['pos'] = data.apply(lambda x: 1 if x['sentiment'] >0  else 0  , axis=1)
data['neg'] = data.apply(lambda x: 1 if x['sentiment'] <0  else 0  , axis=1)
data['zero'] = data.apply(lambda x: 1 if x['sentiment'] ==0  else 0  , axis=1)
```

```python
groupby = data[['listing_id','id']].groupby(['listing_id']).count().reset_index()
dataenglish = data[data['language']=='english']
groupbyenglish = dataenglish[['listing_id','wordcount','charcount','sentiment']].groupby(['listing_id']).mean().reset_index()
groupbyenglishcount = dataenglish[['listing_id','id']].groupby(['listing_id']).count().reset_index()

groupbyenglishsum = dataenglish[['listing_id','pos','neg','zero']].groupby(['listing_id']).sum().reset_index()
```

```python
groupby.columns = ['listing_id','count'] 
groupbyenglishcount.columns = ['listing_id','counteng'] 
```

```python
group1 = groupbyenglishcount.merge(groupbyenglishsum, on ="listing_id")
groupfinal = group1.merge(groupbyenglish, on ="listing_id")
groupfinal['pos'] = groupfinal['pos'] / groupfinal['counteng']
groupfinal['neg'] = groupfinal['neg'] / groupfinal['counteng']
groupfinal['zero'] = groupfinal['zero'] / groupfinal['counteng']
groupfinal['count'] = groupby['count']
```
